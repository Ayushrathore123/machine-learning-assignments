## Split Dataset into X And Y

```python
X=df.drop(labels=["target"],axis=1)
Y=df[["target"]]
```

## perform Standard Scaler on Dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
X_scaled.head()
```

```python
X_scaled.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.9)
X_pca=pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
## plot elbow curve
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
colors = {1:'red', 2:'green', 3:'blue'}
plt.scatter(X_pca[:,0], X_pca[:,1],c=Y['target'].map(colors))
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA with Different class plots')
plt.show()
```

```python
import warnings
```

```python
warnings.filterwarnings("ignore")
```

from sklearn.cluster import KMeans

```python
kmeans=KMeans(n_clusters=3,random_state=42)
```

```python
kmeans.fit(X_pca)
```

```python
preds =kmeans.predict(X_pca)
```

```python
preds
```

```python
colors2 = {0:'red', 1:'green', 2:'blue'}
pred_colors = [colors2[item] for item in preds]
plt.scatter(X_pca[:,0],X_pca[:,1],c=pred_colors)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering of PC1 and PC2')
plt.show()
```
